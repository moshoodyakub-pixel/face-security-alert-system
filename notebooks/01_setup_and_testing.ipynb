## 4. Verify Webcam Access

Let's test if your webcam is accessible and working.

```python
import cv2
import matplotlib.pyplot as plt
from IPython.display import clear_output

print("Testing webcam access...")
print("This will capture a test frame from your webcam.")
print("="*60)

# Try to open webcam
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("❌ Error: Cannot access webcam!")
    print("\nTroubleshooting tips:")
    print("  - Make sure no other application is using the webcam")
    print("  - Check webcam permissions in Windows settings")
    print("  - Try restarting Jupyter Lab")
else:
    # Capture a test frame
    ret, frame = cap.read()
    cap.release()
    
    if ret:
        print("✅ Webcam is working!")
        print(f"   Resolution: {frame.shape[1]}x{frame.shape[0]}")
        print(f"   Channels: {frame.shape[2]}")
        
        # Display the captured frame
        print("\nCaptured frame:")
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        plt.figure(figsize=(10, 8))
        plt.imshow(frame_rgb)
        plt.title("Webcam Test Frame")
        plt.axis('off')
        plt.show()
    else:
        print("❌ Error: Failed to capture frame from webcam")
```

## 5. Test DeepFace Installation

Let's verify DeepFace is working by running a simple face detection test.

```python
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt

print("Testing DeepFace face detection...")
print("This will capture a frame and detect faces in it.")
print("="*60)

# Capture a frame from webcam
cap = cv2.VideoCapture(0)
ret, frame = cap.read()
cap.release()

if ret:
    try:
        # Detect faces using DeepFace
        rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        faces = DeepFace.extract_faces(
            img_path=rgb_frame,
            detector_backend='opencv',
            enforce_detection=False
        )
        
        print(f"✅ DeepFace is working!")
        print(f"   Detected {len(faces)} face(s)")
        
        # Draw boxes on detected faces
        for face in faces:
            facial_area = face['facial_area']
            x, y, w, h = facial_area['x'], facial_area['y'], facial_area['w'], facial_area['h']
            cv2.rectangle(rgb_frame, (x, y), (x+w, y+h), (0, 255, 0), 2)
            cv2.putText(rgb_frame, 'Face', (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)
        
        # Display result
        plt.figure(figsize=(10, 8))
        plt.imshow(rgb_frame)
        plt.title(f"Face Detection Result - {len(faces)} face(s) detected")
        plt.axis('off')
        plt.show()
        
    except Exception as e:
        print(f"❌ Error during face detection: {e}")
else:
    print("❌ Error: Could not capture frame from webcam")
```
